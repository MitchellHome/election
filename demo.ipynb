```python
import numpy as np
import pandas as pd
import re
```

```python
frames = {}
raw_frame= pd.read_csv("raw_ED_2016.csv")
#raw_frame = raw_frame.head(20000)
#raw_frame = raw_frame.drop(['ped_number', 'File', 'DimensionID', 'GNR'], axis=1)
current_dim = raw_frame.iloc[0]["Dimension"]
start = end = 0
#raw_frame.head(200)


#https://open.alberta.ca/dataset/alberta-provincial-electoral-division-profiles-2016-2019-boundaries
```

```python
rows = raw_frame.iterrows()
rows.next()
for i, row in rows:
    if re.match(r'^Total -.*', row["Dimension"]):
        end = i-1
        print "new frame from " + str(start+1) + " to " + str(end)
        if current_dim in frames.keys():
            frames[current_dim] = frames[current_dim].append(raw_frame[start+1:end])
        else:
            frames[current_dim] = pd.DataFrame(raw_frame[start+1:end])
        start = i
        current_dim = row["Dimension"]
        
```
